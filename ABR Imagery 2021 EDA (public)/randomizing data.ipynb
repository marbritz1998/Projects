# Randomzining information form ABR Imagery 2021 Data so I can publicly present my Data Analysis Project

## Importing Modules

```python
import pandas as pd
import numpy as np
from os import devnull
import xlrd
```

```python
#since SO Detail History View is in .Xls format, will need to use devnull and xlrd to open
wb = xlrd.open_workbook('SO Detail History View 2021.Xls', logfile = open(devnull, 'w'))
data = pd.read_excel(wb, thousands = ',')
```

```python
data.head()
```

### Creating a list of unique items and Units, I will use this list to filter out the Unit Price Sell of the original dataset with a new list of Unit Price Sell that is randomaized


```python
#renaming units because the whitespaces are annoying
data = data.rename(columns = {'Units ': 'Units'})
```

```python
#unqiue list of items and their units
items_units = data[['Item ID', 'Units', 'Unit Price Sell']]
```

```python
#making dataframe off of unqiue items 
unique_items_units = pd.DataFrame(items_units.value_counts()).reset_index().rename(columns = {0: 'Unit Price Sell NEW'})
```

```python
#renaming
uiu = unique_items_units
```

### There are items id's that need to be removed, because their amount is dependent on the particular order, checking for duplicates as this list needs to have a one-to-one correspondance

The first thing I need to do is remove the items that could have multiple unit price sells for the same item, like ABR Insurance

This list will be ABR Insurance, Handling and-or cod fee, sublined item, ABR Glass Cut

```python
#removing sublined items, abr insurance, and handling fee
uiu = uiu.loc[~uiu['Item ID'].isin(['ABR INSURANCE', 'Handling and-or cod fee','sublined item', 'ABR Glass Cut'])]
```

```python
#there are duplicates in uiu, getting a list of Item ID's and Units that are duplicated
uiu.loc[uiu.duplicated(subset = ['Item ID', 'Units'])]
```

```python
#keeping first duplicates. note due to how df.duplicated works, If I want to keep only 1, then I have to set the duplicated function == False
uiu = uiu.loc[uiu.duplicated(subset = ['Item ID', 'Units'], keep = 'first') == False]
```

### Creating bin ranges for the randomized prices. I want the prices to be randomized, but I want the magnitude of this new random number to be similar to the magnitude of the original prices, I will then put these random numbers in a new column called Unit Price Sell NEW

```python
#so for bin ranges
# 0-1.99
# 2 - 4.99
# 5 - 9.99
# 10 - 19.99
# 20 - 39.99
# 40 - 79.99
# 80 - 119.99
# 120 - 179.99
# 180 - 229.99
# 230 - 279.99
# 280 - 329.99
# 330 - 379.99
# 380 - 429.99
# 430 - 479.99
# 480 - 529.99
# 530 - 629.99
# 630 - 699.99
# 700 - 999.99
# 1,000 - 1,249.99 Then up $250 up to 4,000.  then anything higher than this will be manually adjusted by me
# I can use the function np.random.uniform() and give a range and desired length
```

```python
bin_ranges = [(0,1.99), (2,4.99), (5, 9.99), (10, 19.99), (20, 39.99), (40, 79.99), (80, 119.99), (120, 179.99), (180, 229.99), (230, 279.99), (280, 329.99), 
              (330, 379.99), (380, 429.99), (430, 479.99), (480, 529.99), (530, 629.99), (630, 699.99), (700, 999.99), (1000, 1249.99), (1250, 1499.99), (1500, 1744.99),
             (1750, 1999.99), (2000, 2249.99), (2250, 2499.99), (2500, 2749.99), (2750, 2999.99), (3000, 3249.99), (3250, 3499.99), (3500, 3749.99), (3750, 3999.99)]
i = 0
while i <= len(bin_ranges):
    bin_low = bin_ranges[i][0]
    bin_high = bin_ranges[i][1]
    uiu.loc[(uiu['Unit Price Sell'] >= bin_low) & (uiu['Unit Price Sell'] <= bin_high), 'Unit Price Sell NEW'] = np.random.uniform(bin_low,bin_high + 0.01, uiu.loc[(uiu['Unit Price Sell'] >= bin_low) & (uiu['Unit Price Sell'] <= bin_high), 'Unit Price Sell'].size)
    i+=1
    if i == len(bin_ranges):
        break
        
```

```python
#great, it worked
uiu.loc[uiu['Item ID'] == 'GLOa1lb']
```

```python
#going to round the Unit Price Sell NEW
uiu['Unit Price Sell NEW'] = uiu['Unit Price Sell NEW'].apply(lambda x: round(x,2))
```

```python
#great, it worked
uiu.loc[uiu['Item ID'] == 'GLOa1lb']
```

### Great, so I now have a list of all the items with randomized unit Price sells that are in the same magnitude as the original data, so now I need to replace the unit price sell in the data with this new unit price sell for all the items

```python
for item, unit in zip(uiu['Item ID'], uiu['Units']):
    if item in set(data['Item ID']):    #for some reason, if this is a list, it will not work as expected, it must be a set
        uiu_item = uiu.loc[(uiu['Item ID'] == item) & (uiu['Units'] == unit)].reset_index(drop = True)  #creating dataframe for each item so I know its index. need this for using df.at[]
        data.loc[(data['Item ID'] == item) & (data['Units'] == unit), 'Unit Price Sell'] = uiu_item.at[0, 'Unit Price Sell NEW']
```

```python
data.loc[data['Item ID'] == '850-400P'].head()
```

## So now I have to re-calculate amount, which is Unit Price Sell * Qty Ship Sell, which I should have the same parameters as the step before so as to only efffect those amount

```python
for row in data.index:
    if data.loc[row,'Price Adj Pct'] == 0:
        data.loc[row, 'Amount'] = data.loc[row, 'Unit Price Sell'] * data.loc[row, 'Qty Ship Sell']
    elif data.loc[row,'Price Adj Pct'] != 0:
        data.loc[row, 'Amount'] = (data.loc[row, 'Unit Price Sell'] * ((100 - data.loc[row, 'Price Adj Pct'])/100)) * data.loc[row, 'Qty Ship Sell']
```

**rounding amount**

```python
data['Amount'] = data['Amount'].apply(lambda x: round(x,2))
```
